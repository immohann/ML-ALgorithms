```
# importing the required libraries.
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```
#generate & plot our own data
X, y = make_blobs(n_samples=350, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:,0], X[:,1],c='red',edgecolors='black')
```

```
#determining clusters using elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```
#finding the centers
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)

#plotting the centers
plt.scatter(X[:,0], X[:,1],c='red',edgecolors='black')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='b',edgecolors='white')
plt.show()
```

```
print(kmeans.cluster_centers_[:, 0])
print(kmeans.cluster_centers_[:, 1])
print("\n\nCluster distribution: \n\n",pred_y)
```
